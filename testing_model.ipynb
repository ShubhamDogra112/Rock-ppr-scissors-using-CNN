```python
import numpy as np
import cv2
from keras_preprocessing import image
from keras.models import load_model
```

```python
try:
    rps = load_model("./rps.h5")
    print("model loaded")

except:
    print("model not found")


def mapper1(index):
    return classes[index]

def mapper2(index):
    return classes[index]


def get_predictions1(img):
    img = image.img_to_array(img)
    img = np.expand_dims(img , axis=0)

    pred = rps.predict(img)
    label = np.argmax(pred[0])
    label = mapper1(label)
    return label


def get_predictions2(img):
    img = image.img_to_array(img)
    img = np.expand_dims(img , axis=0)

    pred = rps.predict(img)
    label = np.argmax(pred[0])
    label = mapper2(label)
    return label

def get_winner(move1 , move2):
    if move1 == move2:
        return "Tie"

    if move1 == "rock":
        if move2 == "scissor":
            return "Player:1"
        if move2 == "paper":
            return "Player:2"

    if move1 == "paper":
        if move2 == "rock":
            return "Player 1"
        if move2 == "scissor":
            return " Player:2"

    if move1 == "scissor":
        if move2 == "paper":
            return "Player:1"
        if move2 == "rock":
            return "Player:2"

classes = {
    0:"paper",
    1:'rock',
    2:'scissor'
}

cap = cv2.VideoCapture(0)


def make_720():
    cap.set(3,1372)
    cap.set(4,720)

make_720()

while True:
    ret,frame = cap.read()
    
    if not ret:
        continue

    frame_gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)

# for player 1
    roi1 = frame[100:500 , 100:500]
    roi_color1 = cv2.cvtColor(roi1 , cv2.COLOR_BGR2RGB)


    img1 = cv2.resize(roi_color1 ,(150,150))
    
    label1 = get_predictions1(img1)
        
    font =cv2.FONT_HERSHEY_SIMPLEX
    
    cv2.putText(frame , 'your move {}'.format(label1) ,(150,50) , font , 1 ,(45,153,255) , 2 ,cv2.LINE_AA)



# for second player

    roi2 = frame[100:500 , 700:1100]
    roi_color2 = cv2.cvtColor(roi2 , cv2.COLOR_BGR2RGB)
    img2 = cv2.resize(roi_color2 ,(150,150))
    
    label2 = get_predictions2(img2)

    cv2.putText(frame , 'your move {}'.format(label2) ,(750,50) , font , 1 ,(45,153,255) , 2 ,cv2.LINE_AA)

    winner = get_winner(label1 , label2)


    cv2.putText(frame , 'Winner is {}'.format(winner) ,(500,600) , font , 1.3 ,(0,0,255) , 2 ,cv2.LINE_AA)


        





    cv2.rectangle(frame ,(100,100) , (500,500) , (0,255,0) ,2 )


    cv2.rectangle(frame ,(700,100) , (1100,500) , (0,255,0) ,2 )




    cv2.imshow("Rock-paper-scissors" , frame)

    pressedKey = cv2.waitKey(1) & 0xFF
    if pressedKey == ord('s'):
        break


cap.release()
cv2.destroyAllWindows()

   

    
```
